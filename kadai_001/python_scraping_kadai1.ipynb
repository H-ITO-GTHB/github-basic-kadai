```python
import requests
from bs4 import BeautifulSoup

# 記事全文を取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/94c907f3d89249c04a346ae5ebbfc8607945ed20'

# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)

# HTMLを解析する
soup = BeautifulSoup(response.text,'html.parser')

paragraphs = soup.select('#uamods > div.article_body p')

# 本文だけを取得する
body_text = paragraphs[1].get_text("\n", strip=True)

# 記事全文を表示する
print(body_text)


```
