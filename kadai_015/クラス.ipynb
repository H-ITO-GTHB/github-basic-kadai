```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 名前と年齢を出力するメソッドを定義する
    def printinfo(self):
        print(self.name)
        print(self.age)

# Humanクラスのインスタンスを作成して変数に代入する
human1 = Human("花子", 16)

# メソッドを呼び出して表示する
human1.printinfo()
```
